```fortran
%fcflags: -w
program test_array_constructor

    integer, allocatable :: a(:)
    real, allocatable    :: b(:)
    real, allocatable    :: c(:)
    integer              :: i
    real, parameter      :: pi = 4*atan(1.0)
    
    a = [(i, i=1, 100)]
    
    array_size = size(a)
    
    b = [(i**2, i=1, 100)]
    
    maximum_value = maxval(b)
    
    c = [(maximum_value * sin(2*pi*i/array_size), i=1, 100)]
       
    !print *, a
    !write (*, "(5X, f0.4)") b
    !write (*, "(5X, f0.10)") c
    
    open(9, file = 'data_straight_line.csv')
    open(8, file = 'data_parabola.csv')
    open(7, file = 'data_sine.csv')
    
    do i = 1, size(a)
        write(9, '(i5)') a(i) 
    end do
    
    do i = 1, size(b)
        write(8, "(5X, f0.4)") b(i)
    end do
    
    do i = 1, size(c)
        write(7, "(5X, f0.4)") c(i)
    end do
end program test_array_constructor
```

```fortran
%fig:figsize=(10, 8)
y_straight_line = np.loadtxt('data_straight_line.csv', unpack = True, delimiter=',') 
y_parabola = np.loadtxt('data_parabola.csv', unpack = True, delimiter=',')
y_sine = np.loadtxt('data_sine.csv', unpack = True, delimiter=',')
x = np.arange(y_straight_line.size)
plt.grid(which='both')
plt.plot(x, y_straight_line*100)
plt.plot(x, y_parabola)
plt.plot(x, y_sine)
plt.title('Test Array Constructors', fontsize = 26)
plt.xlabel('X axis', fontsize = 20)
plt.ylabel('Y axis', fontsize = 20)
plt.savefig('array_constructors_test.png')
```
